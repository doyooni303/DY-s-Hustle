# 배낭문제

```
def knapSack(capa, wt, val, n):
    if n==0 or capa==0:
        return 0

    if wt[n-1] > capa:
        return knapSack(capa, wt, val, n-1)

    else:
        return max(val[n-1] + knapSack(capa-wt[n-1], wt, val, n-1), knapSack(capa, wt, val, n-1))

if __name__ == "__main__":
  n=int(input("Number of items : "))
  weights = list(map(int, input("Weights : ").split()))
  values = list(map(int, input("Values : ").split()))
  bag = int(input("Total Capacity : "))
  print('-'*20)
  print(f'Solution : {knapSack(bag,weights,values,n)}')
```

```
# 부모 집단 내에서 p/100 %만큼 선택 후 crossover 진행
def crossover(popul,p,iteration):
  num_chrom=int(len(popul)*p)
  index_list=list(random.randint(0,len(popul)-1) for _ in range(num_chrom))
  chrom_list=[popul[i] for i in index_list]
  crosspoints=sorted(random.sample(range(1,len(chrom_list[0])-2),2))
  for _ in range(iteration):
    p,q=random.sample(range(0,len(chrom_list)),2)
    a,b = chrom_list[p], chrom_list[q]
    chrom_list[p]=a[:crosspoints[0]]+b[crosspoints[0]:crosspoints[1]]+a[crosspoints[1]:]
    chrom_list[q]=b[:crosspoints[0]]+a[crosspoints[0]:crosspoints[1]]+b[crosspoints[1]:]
  return chrom_list

crossover(population,0.5,5)
```

```
class Node:
  def __init__(self,item,llink,rlink):
    self.item = item
    self.before = llink
    self.next = rlink

class Double:
  def __init__(self):
    self.head = Node(None,None,None)
    self.tail = Node(None,self.head,None)
    self.head.next = self.tail
    self.size = 0

  def is_empty(self):
    return self.size == 0

  def insert_before(self,p,item):
    t = p.before
    n = Node(item,t,p)
    p.before = n
    t.next = n
    self.size += 1

  def insert_after(self,p,item):
    t = p.next
    n = Node(item,p,t)
    t.before = n
    p.next = n
    self.size += 1

  def delete(self,x):
    f = x.before
    r = x.next
    f.next = r
    r.before = f
    self.size -= 1
    return x.item

  def print_list(self):
    if self.is_empty():
      print("List is Empty")
    else:
      p = self.head.next
      while p != self.tail:
        if p.next != self.tail:
          print(p.item, ' <==> ',end='')
        else:
          print(p.item)
        p = p.next

  def search_item(self,item):
    if self.is_empty():
      print("List is Empty")
    else :
      p=self.head
      for i in range(self.size+1):
        if item ==p.item:
          return p
        p=p.next
      
      return None
      
```

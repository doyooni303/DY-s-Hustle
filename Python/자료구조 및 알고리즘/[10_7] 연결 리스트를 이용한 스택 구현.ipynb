```
class Node:
  def __init__(self,item,link):
    self.item = item
    self.next = link

class StackLink:
  def __init__(self):
    self.top = None
    self.size = 0

  def push(self,item):
    #print('top:',self.top)
    self.top = Node(item,self.top)
    self.size += 1
    #print('After push, top :',self.top)

  def pop(self):
    if self.size != 0:
      temp = self.top.item
      self.top = self.top.next
      self.size -= 1
      return temp

  def print_stack(self):
    print('top -->\t', end='')
    p = self.top
    while p:
      if p.next != None:
        print(p.item, ' --> ', end='')
      else:
        print(p.item, end='')
      p = p.next
    print()
```

```
if __name__ == "__main__":
  top = StackLink()
  top.push('apple')
  top.push('mango')
  top.push('cherry')
  top.push('banana')
  top.pop()
  top.print_stack()
```

```
class Node: #이미 이자체로 주소를 의미하는거얀
  def __init__(self, item, link):
    self.item = item
    self.next = link

class QueueLink:
  def __init__(self):
    self.size = 0
    self.front = None
    self.rear = None

  def enqueue(self, item):
    self.np = Node(item,None)
    if self.size == 0:
      self.front = self.np
      self.rear = self.np
    else:
      self.rear.next = self.np
      self.rear = self.np
    self.size += 1
  
  def dequeue(self):
    if self.size != 0:
      delete_item = self.front.item
      self.front = self.front.next
      self.size -= 1
      if self.size == 0:
        self.rear = None
      return delete_item
  
  def queue_print(self):
    p = self.front
    print('front:\t', end=' ')
    while p:
      if p.next != None:
        print(p.item, ' --> ', end='')
        
      else:
        print(p.item, end='')
      p = p.next    
    print('\t:rear')
```

```
if __name__ == "__main__":
  nq = QueueLink()
  nq.queue_print()
  nq.enqueue('apple')
  nq.enqueue('pear')
  nq.enqueue('cherry')
  nq.queue_print()
  nq.dequeue()
  nq.queue_print()
```

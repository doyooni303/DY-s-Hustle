```
class Node:
  def __init__(self, item, left=None, right=None):
    self.item = item # value of Node
    self.left = left # left child of Node
    self.right = right # right child of Node
  
def DFS_stack(T):
    tovisit = [T]
    visited = []
    while tovisit:
      n = tovisit.pop()
      if n.item != None:
        visited.append(n.item)
      
      if n.right != None:
        tovisit.append(n.right)
      
      if n.left != None:
        tovisit.append(n.left)
    return visited

def DFS(t,s):
  tovisit = [s]
  visited = []
  while tovisit:
    u = tovisit.pop()
    visited.append(u)
    for v in reversed(t[u]):
      if v == None:
        continue
      if v not in visited+tovisit:
        tovisit.append(v)
  return visited

if __name__ == "__main__":
  T = Node(1,Node(3,Node(4,None,None),Node(5,None,None)), \
           Node(2, None, Node(7, Node(8,None,None), \
                              Node(6,None,None))))
  print(DFS_stack(T))
  
  item = {100:[200,300],200:[400,500],300:[None,600],400:[None,None], \
          500:[None,None],600:[None,None]}
  
  print(DFS(item,100))
```

```
def odd_even(num):
  if not num.isdecimal() or num.isalpha():
    return f'입력값 오류'
  
  num = int(num)
  if num % 2 == 0:
      return f'{num}은 짝수' #print가 없어도 되네 십탱
  else:
    return f'{num}은 홀수'

if __name__ == "__main__":
  number = input("enter integer >> ")
  print(odd_even(number)) # 함수 호출
```

```
class List_:
  def __init__(self):
    self.lst1 = []
  def add1(self, var):
    self.lst1.append(var)
  def delete1(self,var):
    return self.lst1.remove(var)
  def sort1(self):
    self.lst1.sort() #오름차순 정렬
  def display(self):
    print(f'{self.lst1}')

if __name__ == "__main__":
  var1 = List_()
  var1.add1(10)
  var1.add1(5)
  var1.add1(40);var1.add1(20)
  var1.display()
  var1.sort1()
  var1.display()
```

```
class Node:
  def __init__(self,name,size):
    self.name = name
    self.size = size

if __name__ == '__main__':
  node = []
  for i in range(3):
    node.append(Node(input("name: "),i+1))
  for i in range(3):
    print(node[i].name, node[i].size) #여기서 size는 위치를 의미하나바
```

```
def dfs(x, y):
    #print(f'{x,y}')
    if x <=-1 or x >= n  or y <= -1 or y >= m : #아래의 n,m이랑 맞춰주어야 한다.
        return False
    if map_ice[x][y] == 0:
        # 얼음 = 방문 노드
        map_ice[x][y] = 1
        # 상, 하, 좌, 우 모두 방문해서 함수가 끝나지 않고 모두 방문되었다면
        dfs(x, y+1)
        dfs(x, y-1)
        dfs(x+1, y)
        dfs(x-1, y)
        # 얼음 완성
        return True
    return False

if __name__ == "__main__":
  n,m = map(int,input("enter size >> ").split())
  map_ice = []
  print(f"Input list of 1 or 0 with length {m}")
  for i in range(n):
    print(f"<row{i}> :")
    map_ice.append(list(map(int,input("0 or 1 >> "))))
  answer = 0
  print('\n',np.array(map_ice))
  for x in range(n):
    for y in range(m):
      if dfs(x,y) is True:
        answer +=1
  print('')
  print(f'얼음의 개수 : {answer}개')
```

```
import random
def shopping(demand, budget):
  item=0
  bought=[]
  for i in range(len(demand)):
    budget -=sorted(demand)[i]
    if budget<0:
      break
    bought.append(sorted(demand)[i])
    
    item+=1
  return item, bought
if __name__ == "__main__":
  depart = [random.randint(1,20) for i in range(random.randint(1,10))]
  price = random.randint(1,40)

  print(f"#of Department : {len(depart)}")
  print(f'Demand list : {depart}')
  print(f'Budget : {price}')
  item, bought_list = shopping(depart,price)
  print('------------------Result--------------------')
  print(f"Budget : {price}, Total # of items :{item}, shopping_list :{bought_list} ")
```

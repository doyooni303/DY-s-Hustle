```
import circlecue as circle
```

```
class CircularDeque(circle.CircleQ):#CircleQ가 부모 클래스, Cir- 가 자식 클래스
  def __init__(self,size):
    super().__init__(size) 
#자식함수에 있는 함수를 부모 클래스에 있는 생성자 함수를 그대로 받겠다 
#super()= 자식클래스에도 init이 있지만, 부모클래스안에 있는 init을 실행하겠다 라는 뜻
  def addRear(self,value):
    self.enqueue(value)
  
  def deleteFront(self):
    self.dequeue()
```

```
if __name__=="__main__":
  dq = CircularDeque(8)

  dq.addRear(100) #앞에 한자리 남겨준다
  #dq.addFront(200)
  dq.addRear(300)
  # >> None, 100, 300, ..
  dq.deleteFront()
  # >> None, None, 300, .. 이런 식으로 된다
  dq.display()
```

```
from collections import deque

dq = deque()
print()
item = ['berry','banana']
dq.append('apple')
dq.appendleft('pear')
dq.appendleft(item)
print(dq)
```

```
dq.pop()
dq.popleft()
print(dq)
item = ['cherry','mango']
dq.extend(item)
print(dq)
```

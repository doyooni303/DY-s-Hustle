```
class MaxBHeap:
  def __init__(self,bt):
    self.bt = bt
    self.N = len(bt)-1
#무작위 데이터를 최대 힙으로 구성
#전체 데이터의 중간위치의 자식들 대소 비교
  def create_bheap(self):
    for i in range(self.N//2,0,-1):
      self.down_heap(i)#5,4,3,2,1대입
  
  def insert_bh(self,item):
    self.bt.append(item)
    self.N += 1
    self.up_heap(self.N)

  def delete_bh(self):
    if len(self.bt) == 1:
      print("Tree Empty!1")
    elif len(self.bt) == 2:
      self.bt.pop(-1)
      return self.bt.pop(-1)
    else:
      temp = self.bt[1]
      self.bt[1] = self.bt[self.N]
      #print(f'마지막에 있던 키 값이 첫 번째 키 값으로 지정되었는가 : {self.bt[1]==self.bt[-1]}')
      self.bt.pop(-1) #self.bt.pop(self.N)
      self.N -= 1
      self.down_heap(1)
      return temp
  
  def down_heap(self,i):
    while 2*i <= self.N:
      j = 2*i
      #print(f'(i,j,N)={i,j,self.N}')
      if j==self.N:
        if self.bt[i][0]>=self.bt[j][0]:
          break
        else :
          self.bt[i],self.bt[j] = self.bt[j],self.bt[i]
          break

      #print(f'bt[{j}][0],bt[{j+1}][0]={bt[j][0],bt[j+1][0]}')
      if j < self.N and self.bt[j][0] < self.bt[j+1][0]:
        j = j + 1
        #print(f'j={j}')
        #print(f'bt[{i}][0],bt[{j}][0]={bt[i][0],bt[j][0]}')
        if self.bt[i][0] > self.bt[j][0]:
          break
        #print(f'bt[{i}],self.bt[{j}]={bt[i],self.bt[j]}')
        self.bt[i],self.bt[j] = self.bt[j],self.bt[i]
        #print(f'bt[{i}],self.bt[{j}]={bt[i],self.bt[j]}')
        i = j
      elif j< self.N and self.bt[j][0] >= self.bt[j+1][0]:
        if self.bt[i][0] > self.bt[j][0]:
          break
        else :
          self.bt[i],self.bt[j] = self.bt[j],self.bt[i]
          #print(f'bt[{i}],self.bt[{j}]={bt[i],self.bt[j]}')
          i = j
      else :
        break

  def up_heap(self,i):
    while i> 1:
      if self.bt[i][0] > self.bt[i//2][0]:
        self.bt[i],self.bt[i//2] = self.bt[i//2], self.bt[i]
      i = i//2

  def print_bheap(self):
    for i in range(1,self.N+1):
      print(f'[{self.bt[i][0]:3d},{self.bt[i][1]}]',end='')
      print('\nSize of Heap : ', self.N)
  
  def sort_heap(self):
    sort_max = []
    while len(self.bt) > 1:
      sort_max.append(self.delete_bh())
    return sort_max
```

```
import heapq
def heapq_module(value):
  data = []
  result = []
  for s1,s2 in value:
    heapq.heappush(data, [s1,s2])
  while len(data) > 0:
    a,b = heapq.heappop(data)
    result.append([a,b])
  return result

if __name__ == "__main__":
  value = [[90,'melon'],[25,'pear'],[70,'lime'],[40,'cherry'],[60,'grape'] \
  ,[20,'orange'],[30,'apricot'],[10,'lemon'],[75,'kiwi'],[45,'mango']]
  result = heapq_module(value)
  print(result)
```

```
def heapq_min_module(value):
  data = []
  result = []
  for s1 in value:
    heapq.heappush(data, s1)
  while len(data) > 0:
    a = heapq.heappop(data)
    result.append(a)
  return result

```

```
import heapq
def heapq_max_module(works):
    data = []
    result =[]
    for value in works:
        heapq.heappush(data,-value)
    while len(data)>0:
      a=heapq.heappop(data)
      result.append(-a)
    return result
```

# 우선순위큐 적용 알고리즘 문제 1

```
import heapq
```

```
def heapq_min_module(value):
  data = []
  result = []
  for s1 in value:
    heapq.heappush(data, s1)
  while len(data) > 0:
    a = heapq.heappop(data)
    result.append(a)
  return result

```

```
if __name__ == "__main__":
  value = [1,2,3,9,10,12]
  for i in value:
    print(i,end=' ')
  print()
  result = heapq_min_module(value)
  
  k=int(input('input the threshold : '))
  
  scoville=0
  i=0
  while result[0] < k:
    i+=1
    first=heapq.heappop(result)
    #print(first)
    second=heapq.heappop(result)
    #print(second)
    scoville=first+second*2
    #print(scoville)
    heapq.heappush(result,scoville)
    result=heapq_min_module(result)
    #print(result)

  print(f'{result,i}')
```

# 우선순위큐 적용 알고리즘 문제 2

```
def heapq_max_module(values):
    data = []
    result =[]
    for value in values:
        heapq.heappush(data,-value)
    while len(data)>0:
      a=heapq.heappop(data)
      result.append(-a)
    return result
```

```
def fatigue(n, works):
    if sum(works) < n: 
      return print("I can handle it!")
    
    left=heapq_max_module(works)
    for i in range(n):
        update = heapq.heappop(left)
        value = update - 1
        heapq.heappush(left,value)
        left=heapq_max_module(left)

        
    works = [work**2 for work in left]
    return print(left,n,sum(works))

```

```
Works = input('enter work load >> ').split()
for i in range(len(Works)):
 Works[i]=int(Works[i])
N = int(input('enter remaining work times >> '))

if len(Works)>200 or min(Works)<=0 or max(Works)>500 or N>10000:
   print("""You've got the wrong input. """)

fatigue(N,Works)
```

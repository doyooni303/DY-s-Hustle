```
def gradient_descent(f, init_x, lr, step_num):
    x = init_x
    x_history = []

    for i in range(step_num):
        x_history.append( x.copy() )

        grad = numerical_gradient(f, x)
        x -= lr * grad

    return x, np.array(x_history)

#잊지 말아야 할 것은 w에 따른 손실함수 변화인 것이고, dL/dw가 0이 되는 w를 찾아주는 것이 우리 목표

def function_2(x):
    return np.sum(x**2) #x가 고차원이어도 상관없음


init_x = np.array([-3.0, 4.0, 5, 6])    

lr =0.01
step_num = np.arange(1,75+1,1)

#이건 iteration에 변화를 줬을 때
#for alpha in lr:
#  print("------------------------------------------------------------------------------------------------")
#  for iteration in step_num:
#    opt_x, x_history = gradient_descent(function_2, init_x, lr=alpha, step_num= iteration)
#    print("{0}: learningrate:{1:.5f},opt_x:{2}".format(iteration,alpha,opt_x))
optimal=np.array([])
optimal=optimal.reshape(-1,4)
for i in step_num:
  opt_x, x_history = gradient_descent(function_2, init_x, lr=lr, step_num= i)
  optimal = np.append(optimal, opt_x.reshape(-1,4))
  print("{0}: {1}".format(i,opt_x))

optimal
```

Operations

There are multiple syntaxes for operations. \

In the following example, we will take a look at the addition operation.

Addition: syntax 1



```
y=torch.rand(5,3)
z = torch.rand(5, 3)
print(y)
print(z)
print(y + z)
```

```
print('y=',y)
print('z=',z)
print(torch.add(z, y)== (z+y))
print(f'torch.mul(z, y)=\n{torch.mul(z, y)}')
print(f'torch.sub(y,z)=\n{torch.sub(y,z)}')
print(f'torch.sub(z,y)=\n{torch.sub(z,y)}')
```

Addition: providing an output tensor as argument



```
result = torch.rand(5, 3)
print('result=\n',result)

torch.add(z, y)
print(result)

torch.add(z, y, out=result) #result에 값을 저장함 
print(result)
```

Addition: in-place



```
# adds z to y
y.add_(z) #= y+=z
print(y)
```

<div class="alert alert-info"><h4>Note</h4><p>Any operation that mutates a tensor in-place is post-fixed with an ``_``.
    For example: ``x.copy_(y)``, ``x.t_()``, will change ``x``.</p></div>

You can use standard NumPy-like indexing with all bells and whistles!



```
x=torch.zeros(5,3,dtype=torch.int32)
print(x)
print(y)
print(z.copy_(y)) #z를 y값으로 복사하는것임
print(z) # 그럼 z가 y로 변함
print(z.copy_(x))
print(y[:, 0]) # 0차원 값들
```

Resizing: If you want to resize/reshape tensor, you can use ``torch.view``:



```
x = torch.randn(12)
print(x,x.size())
y = x.view(1,12)
print(y,y.size())
y = x.view(2,6)
print(y,y.size())
y = x.view(3,4)
print(y,y.size())
y = x.view(4,3)
print(y,y.size())
y = x.view(6,2)
print(y,y.size())

z = x.view(-1, 6)  # the size -1 is inferred from other dimensions
print(z)
print(x.size(), y.size(), z.size())
```

If you have a one element tensor, use ``.item()`` to get the value as a
Python number



```
x = torch.randint(1,10,(1,)) #정수 뽑아낼 때는 위-아래 범위 정해주고 튜플로 사이즈를 입력해야한다.
print(x)
print(type(x.item()))
```

NumPy Bridge
------------

Converting a Torch Tensor to a NumPy array and vice versa is a breeze.

The Torch Tensor and NumPy array will share their underlying memory
locations (if the Torch Tensor is on CPU), and changing one will change
the other.

Converting a Torch Tensor to a NumPy Array




```
a = torch.ones(5,dtype=torch.long)
print(a,type(a),a.size())
b = a.numpy()
print(b,type(b),b.size)
```

```
a.add_(1)
print(a)
print(b) #b도 같이 저장 되어서 반영되네 십탱구리
print(b+1,a) #b를 바꾼다고 a가 바뀌지는 않음
```

Converting NumPy Array to Torch Tensor

See how changing the np array changed the Torch Tensor automatically



```
import numpy as np
a = np.ones(5)
b = torch.from_numpy(a)
print(a,type(a))
print(b,type(b))
np.add(a, 1, out=a)
print(a)
print(b) #반대로 a를 변화시키면 b도 같이 변하게 되는구먼
```

GPU 연산을 위한 자료형도 있습니다. 예를 들어 torch.cuda.FloatTensor

```
x = torch.FloatTensor([[1, 2], [3, 4]])
y = torch.FloatTensor([[5, 6], [7, 8]])
```

```
print(torch.cat([x, y], dim=0)) # cat= concatnate
print(torch.cat([ y,x], dim=0)) # 열의 길이가 같아야 함

print(torch.cat([x, y], dim=1))
print(torch.cat([ y,x], dim=1)) # 행의 길이가 서로 같아야 함
```

```
x = torch.FloatTensor([1, 4])
y = torch.FloatTensor([2, 5])
z = torch.FloatTensor([3, 6])

print(torch.stack([x, y, z])) # 꼭 []로 마지막을 묶어 줘야 함 / 이렇게 되면 그냥 2차원 배열이 3개의 행으로 구성됨 so (3,2) 가 되지
```

```
x=torch.FloatTensor([1])
y=torch.FloatTensor([4])
z=torch.FloatTensor([5])

print(torch.stack([x,y,z]))
print(torch.stack([x,y,z]).size()) # 그냥 행으로 쌓게 됨. 그래서 열의 길이가 같아야 함
```

```
x = torch.FloatTensor([1, 4])
y = torch.FloatTensor([2, 5])
z = torch.FloatTensor([3, 6])

print(torch.stack([x, y, z],dim=1))
#(1,2)를 (2,1)로 바꾸고 열로 쌓아버림
```

```
x=torch.FloatTensor([1])
y=torch.FloatTensor([4])
z=torch.FloatTensor([5])

print(torch.stack([x,y,z],dim=1))
print(torch.stack([x,y,z],dim=1).size())
# 열로 쌓음
```

그런데 연산 뒤에 _를 붙이면 기존의 값을 덮어쓰기 합니다.

```
x = torch.FloatTensor([[1, 2], [3, 4]])
print(x.mul(2)) # 곱하기 2를 수행한 결과를 출력
print(x)
print(x.mul_(2)) # 곱하기 2를 수행한 결과를 출력
print(x)
```

```
class Calculator:
    def __init__(self): # 객체 생성 시 호출될 때 실행되는 초기화 함수. 이를 생성자라고 한다.
        self.result = 0

    def add(self, num): # 객체 생성 후 사용할 수 있는 함수.
        self.result += num
        return self.result
```

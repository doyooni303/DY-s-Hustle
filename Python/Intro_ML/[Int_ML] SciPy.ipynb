```
from scipy import sparse
from scipy.sparse import coo_matrix
# 대각선 원소는 1이고 나머지는 0인 2차원 NumPy 배열을 만듭니다.
eye = np.eye(4)
print("NumPy 배열:\n", eye)
```

```
# NumPy 배열을 CSR 포맷의 SciPy 희박 행렬로 변환합니다. 0이 아닌 원소만 저장됩니다.
sparse_matrix = sparse.csr_matrix(eye)
sparse_matrix
print("\nSciPy의 CSR 행렬:\n", sparse_matrix)
```

```
row = np.array([0, 3, 1, 0]) 
col = np.array([0, 2, 3, 2]) 
data = np.array([4, 5, 7, 9]) 
coo_matrix((data, (row, col)), shape=(4, 4)).toarray() #.toarray()는 sparse matrix를 실제 배열로 매트리스로 보여주는 거구먼
```

```
data = np.ones(4) #[1,1,1,1]

np.arange(100) # 0부터 n까지 연속 수열을 만들어주는 것이구먼
row_indices = np.arange(4) #[0,1,2,3]
col_indices = np.arange(4) #[0,1,2,3]
eye_coo=coo_matrix((data, (row_indices, col_indices)))
print(eye_coo)#불필요한 공백을 두지 않는 것이고
eye_coo_toarray = coo_matrix((data, (row_indices, col_indices))).toarray()
print(eye_coo_toarray)
 # 좋았어
```

```
from sklearn.datasets import load_iris
iris_dataset = load_iris()
print("iris_dataset의 Key:\n", iris_dataset.keys())
print(iris_dataset['DESCR'])
```

```
type(iris_dataset)
```

```
print("타깃의 이름:", iris_dataset['target_names'],"\n")
print("특성의 이름:\n", iris_dataset['feature_names'],"\n")
print("data의 타입:", type(iris_dataset['data']),"\n")
print("data의 크기:", iris_dataset['data'].shape,"\n")
print("data의 처음 다섯 행:\n", iris_dataset['data'][:5],"\n")
print("target의 타입:", type(iris_dataset['target']),"\n")
print("targetnames의 타입:", type(iris_dataset['target_names']))
display(iris_dataset.target_names)
display(iris_dataset.target)
```

```
from sklearn.model_selection import train_test_split
train_test_split #<function sklearn.model_selection._split.train_test_split>

X_train, X_test, y_train, y_test = train_test_split(iris_dataset['data'], iris_dataset['target'], random_state=0)
```

```
print("X_train 크기:", X_train.shape)
print("y_train 크기:", y_train.shape)
```

```
print("X_test 크기:", X_test.shape)
print("y_test 크기:", y_test.shape)
```

```
# X_train 데이터를 사용해서 데이터프레임을 만듭니다.
# 열의 이름은 iris_dataset.feature_names에 있는 문자열을 사용합니다.
iris_dataframe = pd.DataFrame(X_train, columns=iris_dataset.feature_names)
# 데이터프레임을 사용해 y_train에 따라 색으로 구분된 산점도 행렬을 만듭니다.

pd.plotting.scatter_matrix(iris_dataframe, c=y_train, figsize=(15, 15), marker='o',
                           hist_kwds={'bins': 20}, s=60, alpha=.8, cmap=mglearn.cm3)
plt.show() # 책에는 없음
```

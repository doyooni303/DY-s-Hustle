```
from sklearn.datasets import load_iris
iris_dataset = load_iris()
print("iris_dataset의 Key:\n", iris_dataset.keys())
print(iris_dataset['DESCR'])
```

```
from sklearn.model_selection import train_test_split
train_test_split #<function sklearn.model_selection._split.train_test_split>

X_train, X_test, y_train, y_test = train_test_split(iris_dataset['data'], iris_dataset['target'], random_state=0)
```

```
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
```

```
knn.fit(X_train, y_train)
```

```
X_new = np.array([[5, 2.9, 1, 0.2]])
print("X_new.shape:", X_new.shape)

prediction = knn.predict(X_new)
print("예측:", prediction)
print("예측한 타깃의 이름:", 
       iris_dataset['target_names'][prediction])
```

```
y_pred = knn.predict(X_test)
print("테스트 세트에 대한 예측값:\n", y_pred)
print("테스트 세트에 대한 실제값:\n", y_test)
np.where(y_pred==y_test) # 조건에 맞는 배열의 index 찾는 방법
```

```
print("테스트 세트의 정확도: {:.2f}".format(np.mean(y_pred == y_test))) #37/38
```

```
#neighbor 수에 따른 정확도 

N=np.arange(0,4,1)
knn_name=[]
knn='Knn'

Y_pred=[]

for n in N:
    knn_name.append(knn+str(N[n]))
    Y_pred.append('Y_pred'+str(N[n]))
print(knn_name)
print(Y_pred)
```

```
for n in N:
    from sklearn.neighbors import KNeighborsClassifier
    knn_name[n] = KNeighborsClassifier(n_neighbors=n+1)
    knn_name[n].fit(X_train, y_train)
    Y_pred[n] = knn_name[n].predict(X_test)
    print("K={0}".format(n+1))
    print("테스트 세트에 대한 예측값:\n", Y_pred[n])
    print("테스트 세트에 대한 실제값:\n",y_test)
    print("테스트 세트의 정확도: {:.2f}\n\n".format(np.mean(Y_pred[n] == y_test)))
```

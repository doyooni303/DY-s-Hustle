```
from sklearn.datasets import load_breast_cancer
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
 cancer = load_breast_cancer()

for i in range(0,50):
   X_train, X_test, y_train, y_test = train_test_split(
  cancer.data, cancer.target, stratify=cancer.target, random_state=i)
  logreg = LogisticRegression(max_iter=j).fit(X_train, y_train)
  print("iteration:{0}".format(i+1))
  print("훈련 세트 점수: {:.5f}".format(logreg.score(X_train, y_train)))
  print("테스트 세트 점수: {:.5f}\n".format(logreg.score(X_test, y_test)))
```

```
logreg100 = LogisticRegression(C=100, max_iter=10000).fit(X_train, y_train)
print("훈련 세트 점수: {:.3f}".format(logreg100.score(X_train, y_train)))
print("테스트 세트 점수: {:.3f}\n".format(logreg100.score(X_test, y_test)))
logreg = LogisticRegression( max_iter=10000).fit(X_train, y_train)
print("훈련 세트 점수: {:.3f}".format(logreg.score(X_train, y_train)))
print("테스트 세트 점수: {:.3f}\n".format(logreg.score(X_test, y_test)))
logreg001 = LogisticRegression(C=0.01, max_iter=10000).fit(X_train, y_train)
print("훈련 세트 점수: {:.3f}".format(logreg001.score(X_train, y_train)))
print("테스트 세트 점수: {:.3f}".format(logreg001.score(X_test, y_test)))
```

```
plt.figure(figsize=(24,16))
plt.plot(logreg100.coef_.T, '^', label="C=100",markersize=30)
plt.plot(logreg.coef_.T, 'o', label="C=1",markersize=30)
plt.plot(logreg001.coef_.T, 'v', label="C=0.001",markersize=30)
plt.xticks(range(cancer.data.shape[1]), cancer.feature_names, rotation=60,fontsize=20) #cancer.data.shape=(560,30) 따라서 range(~[1])은 (0,30)이 됨.
xlims = plt.xlim() # (0, 2) 임
plt.hlines(0, xlims[0], xlims[1]) # 수평선을 y=0을 (0,0) 부터 (2,0)까지 그리겠다 이것이지
plt.xlim(xlims)
plt.ylim(-5, 5)
plt.yticks(fontsize=20)
plt.xlabel("Features",fontsize=50)
plt.ylabel("Coefficients",fontsize=50)
plt.legend(loc=0,fontsize=50)
```

```
plt.figure(figsize=(15,10))
for C, marker in zip([0.001, 1, 100], ['o', '^', 'v']):
    lr_l1 = LogisticRegression(solver='liblinear', C=C, penalty="l1", max_iter=1000).fit(X_train, y_train)
    print("C={:.3f} 인 l1 로지스틱 회귀의 훈련 정확도: {:.2f}".format(
          C, lr_l1.score(X_train, y_train)))
    print("C={:.3f} 인 l1 로지스틱 회귀의 테스트 정확도: {:.2f}".format(
          C, lr_l1.score(X_test, y_test)))
    plt.plot(lr_l1.coef_.T, marker, label="C={:.3f}".format(C),markersize=10)

plt.xticks(range(cancer.data.shape[1]), cancer.feature_names, rotation=60,fontsize=15)
xlims = plt.xlim()
plt.hlines(0, xlims[0], xlims[1])
plt.xlim(xlims)
plt.xlabel("Features",fontsize=15)
plt.ylabel("Coefficients",fontsize=15)
plt.yticks(fontsize=20)
plt.ylim(-5, 5)
plt.legend(loc=3,fontsize=10)
```

```
from sklearn.datasets import make_blobs

X, y = make_blobs(random_state=33)
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
plt.xlabel("특성 0")
plt.ylabel("특성 1")
plt.legend(["클래스 0", "클래스 1", "클래스 2"])
```

```
from sklearn.svm import LinearSVC
linear_svm = LinearSVC().fit(X, y)
print("계수 배열의 크기: ", linear_svm.coef_.shape)
print("절편 배열의 크기: ", linear_svm.intercept_.shape)
```

```
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
line = np.linspace(-15, 15) #-15,15 포함 기본 50개의 수열 생성
for coef, intercept, color in zip(linear_svm.coef_, linear_svm.intercept_,
                                  mglearn.cm3.colors):
    plt.plot(line, -(line * coef[0] + intercept) / coef[1], c=color)
plt.ylim(-10, 15)
plt.xlim(-10, 8)
plt.xlabel("Features 0")
plt.ylabel("Features 1")
plt.legend(['Class 0', 'Class 1', 'Class 2', 'Class 0 Border', 'Class 1 Border',
            'Class 2 Border'], loc=(1.01, 0.3))
```

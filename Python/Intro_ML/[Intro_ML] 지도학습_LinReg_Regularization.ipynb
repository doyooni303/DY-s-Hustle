```
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
X, y = mglearn.datasets.make_wave(n_samples=1000)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=100)

lr = LinearRegression().fit(X_train, y_train)

print("lr.coef_:", lr.coef_)
print("lr.intercept_:", lr.intercept_)
```

```
print("훈련 세트 점수: {:.2f}".format(lr.score(X_train, y_train)))
print("테스트 세트 점수: {:.2f}".format(lr.score(X_test, y_test)))
```

```
X, y = mglearn.datasets.load_extended_boston()

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=33)
boston_lr = LinearRegression().fit(X_train, y_train)
```

```
from sklearn.datasets import load_boston
boston = load_boston()
boston.feature_names
```

```
print("훈련 세트 점수: {:.2f}".format(boston_lr.score(X_train, y_train)))
print("테스트 세트 점수: {:.2f}".format(boston_lr.score(X_test, y_test)))
print("사용한 특성의 개수:{0}\n".format(np.sum(boston_lr.coef_ != 0)) )
#얘는 직선 모양이 맞는거지
```

```
#Ridge Model
from sklearn.linear_model import Ridge

for i in range(0,10):
  ridge = Ridge(alpha=i).fit(X_train, y_train)
  print("alpha={}".format(i))
  print("훈련 세트 점수: {:.2f}".format(ridge.score(X_train, y_train)))
  print("테스트 세트 점수: {:.2f}".format(ridge.score(X_test, y_test)))
  print("사용한 특성의 개수:{0}\n".format(np.sum(ridge.coef_ != 0)) )
```

```
from sklearn.linear_model import Lasso
alpha_list=np.arange(0,5,0.01)
coef_list=[]

for i in alpha_list:
  lasso = Lasso(alpha=i).fit(X_train, y_train)
  coef_list.append(lasso.coef_)
```

```
plt.figure(figsize=(30,50))
plt.plot(alpha_list,coef_list)
plt.legend(labels=boston.feature_names,fontsize=35)
plt.hlines(y=0,xmin=np.min(alpha_list),xmax=np.max(alpha_list),colors='black',linestyles='-')
plt.ylim(np.min(coef_list),100)
plt.grid()
plt.show()
```
